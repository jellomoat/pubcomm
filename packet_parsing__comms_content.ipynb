## Convert image to text

```python
def get_all_files(dir_path):
    """
    Lists all files in specified directory.

    `dir_path`: path to target directory
    """
    dir_list = os.listdir(dir_path)

    print("Files and directories in '", dir_path, "' :")
    print(dir_list)
    return dir_list
```

```python
def get_text_from_pg_img(img_path):
    return pytesseract.image_to_string(Image.open(img_path))
```

```python
pdf_folder = "./pubcomm/pdf_images_2/"
```

## Convert PDF to text
Doing this together because of sorting issues when retrieving images from the image dir

```python
test_fpath_2 = "./raw_pdfs/7_11_2023; CLK - Communication; Public; ; REGULAR; Supplemental Communicatons 2.pdf"
test_fpath_3 = "./raw_pdfs/7_11_2023; CLK - Communication; Public; ; REGULAR; Supplemental Communications 3.pdf"
```

```python
pdf_folder = "./pdf_images_3/"

with open("./comms_content_3.txt", "w") as f:
    pages = convert_from_path(test_fpath_3, 350)

    for i in range(len(pages)):
        # Save pages as images in the pdf
        img_path = f"page{i}.jpg"
        pages[i].save(pdf_folder + img_path, 'JPEG')
        page_text = get_text_from_pg_img(pdf_folder + img_path)
        f.write(page_text)
        f.write("\n----\n")
```

```python
with open("./comms_content_3.txt", "r") as f:
    lines = f.read()
```

```python
content = re.split(r"----\n([0-9]+)\n", lines)[1:]
```

```python
content_dicts_list = []

for i in range(len(content)):
    if i % 2:
        m = re.search(r"\s*Benado,*\s*Tony\n*([\s\S]*)", content[i])
        if m:
            content_dict = {
                "comms_num": int(content[i-1]),
                "comms_text": m.group(1).strip()
            }
            content_dicts_list.append(content_dict)

for d in content_dicts_list:
    print(d)
    print("----")
```
