```python
import re
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://berkeleyca.gov/city-council-regular-meeting-eagenda-july-11-2023"
```

```python
resp = requests.get(url)
```

```python
soup = BeautifulSoup(resp.text, "html")
```

```python
field_items = soup.select("div.field-items > div.field-item > div > p")
```

```python
agenda_items = []
for item in field_items:
    if item.select("strong > a"):
        try:
            item_dict = {
                "item_num": re.search(r"([0-9]*).", item.select_one("strong").text).group(1),
                "item_name": item.select_one("strong > a").text
            }
            content = re.search(r"Recommendation:\s*(.*)Financial Implications:\s*(.*)\nContact:\s*(.*)", item.text)
            item_dict["rec"] = content.group(1)
            item_dict["financial_implications"] = content.group(2)
            item_dict["contact"] = content.group(3)
            agenda_items.append(item_dict)
        except:
#             print(item)
            pass
print(agenda_items)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(agenda_items)
```

```python
df.to_csv("./agenda_items_071123b.csv", index="item_num")
```
